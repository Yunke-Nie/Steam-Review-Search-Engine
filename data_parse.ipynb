```python
import pandas as pd
import numpy as np
from concurrent.futures import ThreadPoolExecutor
from textblob import TextBlob
```

```python
data = pd.read_csv('steam_reviews.csv')
data.head()
```

```python
def classify_sentiment(text):
    # Ensure that the text is a string
    if isinstance(text, str):
        analysis = TextBlob(text)
        if analysis.sentiment.polarity > 0:
            return 'Positive'
        elif analysis.sentiment.polarity == 0:
            return 'Neutral'
        else:
            return 'Negative'
    else:
        # Return a default value or a placeholder if the text is not a string
        return 'Unknown'
```

```python
def apply_sentiment_multithreaded(df, column_name):
    with ThreadPoolExecutor() as executor:
        sentiments = list(executor.map(classify_sentiment, df[column_name]))
    return sentiments
```

```python
data['sentiment'] = apply_sentiment_multithreaded(data, 'review')
data.head()
```

```python
data.to_csv('dataset_sentiment.csv', index=False)
```
